```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
def cartoonify(image, arguments=0):
    
    imgBlur = cv2.GaussianBlur(image,(11,11),0)
    mask = np.zeros(image.shape,dtype = np.uint8)
    pencilSketchImage = pencilSketch(image)
    mask[pencilSketchImage == 0] = 1
    cartoonImage = (1-mask)*imgBlur + mask*pencilSketchImage
    return cartoonImage
```

```python
def pencilSketch(image, arguments=0):
    
    imgGray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    imgGrad = cv2.Laplacian(imgGray,cv2.CV_32F,ksize=7)
    
    cv2.normalize(imgGrad,imgGrad,0,1,norm_type = cv2.NORM_MINMAX)
    imgGrad = np.uint8(imgGrad*255)
    
    v = np.median(imgGrad)
    thresh = v + int(.04 * v)
    
    _,imgThresh = cv2.threshold(imgGrad,thresh,255,cv2.THRESH_BINARY)
    imgThresh = 255 - imgThresh
    
    pencilSketchImage = cv2.merge((imgThresh,imgThresh,imgThresh))
    return pencilSketchImage
```

```python
imagePath = "./imgs/portrait.jpg"
image = cv2.imread(imagePath)

hRatio = 960 / image.shape[0] 
wRatio = 640 / image.shape[1] 


if max(hRatio, wRatio) < 1:
    interpolation = cv2.INTER_AREA
else:
    interpolation = cv2.INTER_CUBIC

image = cv2.resize(image, tuple([0,0]), fx=max(hRatio, wRatio), fy=max(hRatio, wRatio), interpolation=interpolation)
```

```python
cartoonImage = cartoonify(image)
pencilSketchImage = pencilSketch(image)
```

```python
plt.figure(figsize=[20,10])
plt.subplot(131);plt.imshow(image[:,:,::-1]);plt.axis('off');
plt.subplot(132);plt.imshow(cartoonImage[:,:,::-1]);plt.axis('off');
plt.subplot(133);plt.imshow(pencilSketchImage[:,:,::-1]);plt.axis('off')
```
